```python
# 笨方法
class Solution(object):
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        def search_val(root, val_list):
            if not root:
                return val_list
            val_list.insert(0,root.val)
            return search_val(root.next,val_list)
        l1_val = search_val(l1,[])
        l2_val = search_val(l2,[])
        if len(l1_val) > len(l2_val):
            min_len = len(l2_val)
            res_val = l1_val
            target = l2_val
        else:
            min_len = len(l1_val)
            res_val = l2_val
            target = l1_val
        for i in range(min_len):
            if res_val[i] >= 10:
                if i+1 == len(res_val):
                    res_val.append(1)
                else:
                    res_val[i + 1] += 1
                res_val[i] = res_val[i]%10
            res_val[i] = res_val[i] + target[i]
            if res_val[i] >= 10:
                if i+1 == len(res_val):
                    res_val.append(1)
                else:
                    res_val[i + 1] += 1
                res_val[i] = res_val[i]%10
        for i in range(min_len, len(res_val)):
            if res_val[i] >= 10:
                if i+1 == len(res_val):
                    res_val.append(1)
                else:
                    res_val[i + 1] += 1
                res_val[i] = res_val[i]%10
        head = ListNode(None)
        pre = head
        for i in res_val[::-1]:
            new_node = ListNode(i)
            pre.next = new_node
            pre = new_node
        return head.next
                
```

```python
class Solution(object):
    def addTwoNumbers(self, l1, l2):
        s1, s2 = [], []
        while l1:
            s1.append(l1.val)
            l1 = l1.next
        while l2:
            s2.append(l2.val)
            l2 = l2.next
        ans = None
        carry = 0
        while s1 or s2 or carry != 0:
            a = 0 if not s1 else s1.pop()
            b = 0 if not s2 else s2.pop()
            cur = a+b+carry
            carry = cur//10
            cur %= 10
            curnode = ListNode(cur)
            curnode.next = ans
            ans = curnode
        return ans
```
